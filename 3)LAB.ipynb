```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("casasboston.csv")
```

```python
print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
```

```python
print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist()
plt.title("Grafico de dispercion por indice de criminalidad")
plt.xlabel("Valor")
plt.ylabel("")
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=1)
plt.title("Grafico de dispercion por indice de criminalidad")
plt.xlabel("Valor de mediano")
plt.ylabel("Indice de crimen")
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
```

```python
valor_por_ciudad.head(10).plot.barh()
plt.title("Grafico de dispercion por indice de criminalidad")
plt.xlabel("Valor")
plt.ylabel("Ciudad")
plt.show()
```
